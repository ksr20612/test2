```python
import tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
(x_train,y_train), (x_test,y_test)=tensorflow.keras.datasets.boston_housing.load_data(path='boston_housing.npz')
```

```python
x_train.shape  # 404개의 데이터. 13개의 영역
```

```python
y_train.shape # 404개의 데이터, 1개의 값
```

```python
x_test.shape
```

```python
y_test
```

```python
model=Sequential()
model.add(Dense(1, input_shape=[13]))
model.compile(optimizer='rmsprop', loss='mse')
model.summary()
```

```python
model.fit(x_train, y_train, epochs=1000) #x train 에는 13개의 칼럼이, y train 에는 1개의 칼럼이 들어있음 #fit="train"
```

```python
model.evaluate(x_test, y_test) #evaluate="test", 그냥 한번 테스트해보기
```

```python
model.predict(np.array([0,1,50,1,0,5,5,5,10,20,30,15,10])) #predict="infer", 예측하기 
```
